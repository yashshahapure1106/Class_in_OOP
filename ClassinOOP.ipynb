#### Creating user defined class

```python
class Person:
    
    """
    This class defines a Person
    """
    def __init__(self, name, age, gender, occupation):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation
        
    def introduce(self):
        """
        This function will introduce the person
        """
        print(f'Hi there my name is {self.name}')
        print(f'I am currently {self.age} years old')
        print(f'I am a {self.gender}')
        print(f'I work as a {self.occupation}')
    
    def work(self, company):
        """
        This function talks about work of person
        """
        print(f'{self.name} works at {company}')
```

```python
p1 = Person('Yashraj', 24, 'Male', 'Data Scientist')
```

### Bank Account class example

```python
class Account:
    
    def __init__(self, ac_no, name, balance):
        self.ac_no = ac_no
        self.name = name
        self.balance = balance
        
    def deposit(self, amt):
        self.balance = self.balance + amt
        print(f'Money of {amt} INR credited into account : {self.ac_no}')
        print(f'Current Balance : {self.balance} INR')
        
    def withdraw(self, amt):
        if self.balance >= amt:
            self.balance = self.balance - amt
            print(f'Money of {amt} INR debited from your account : {self.ac_no}')
            print(f'Current Balance : {self.balance} INR')
        else:
            print(f'Insufficient Balance in account : {self.ac_no}')
            print(f'Current Balance : {self.balance} INR')
            
    def transfer(self, amt, tr_ac):
        if self.balance >= amt:
            self.balance = self.balance - amt
            tr_ac.balance = tr_ac.balance + amt
            print(f'Money of {amt} INR debited from your account no {self.ac_no} to {tr_ac.ac_no}')
            print(f'Current Balance : {self.balance} INR')
        else:
            print(f'Insufficient Balance in account : {self.ac_no}')
            print(f'Current Balance : {self.balance} INR')   
```
